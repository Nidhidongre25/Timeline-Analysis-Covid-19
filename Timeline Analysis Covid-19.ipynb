```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(16,8)})
sns.lineplot(x='date', y='total_cases', data=india_case)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.bar('total_cases','total_deaths', data=india_case)
plt.xticks(rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10 , 6))
ax.scatter(india_case["total_cases"] , india_case["total_deaths"])
ax.set_xlabel('total_cases')
ax.set_ylabel('total_deaths')
plt.show()
```
